```python
# Notebook 1: convert_html_to_excel.ipynb

import os
import pandas as pd

input_dir = r"output\excels"
converted_dir = r"output\html_xlsx_converted"
os.makedirs(converted_dir, exist_ok=True)

def convert_html_xls_to_xlsx(file_path, output_dir):
    try:
        tables = pd.read_html(file_path, header=None)
        if not tables:
            print(f"⚠️ No tables found in {file_path}")
            return

        df = tables[0]
        df = df.dropna(how='all').dropna(axis=1, how='all')

        # Flatten MultiIndex columns if needed
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = [' '.join([str(c) for c in col if str(c) != 'nan']).strip() for col in df.columns]

        # Save as real .xlsx
        output_file = os.path.join(output_dir, os.path.basename(file_path).replace(".xls", ".xlsx"))
        df.to_excel(output_file, index=False)
        print(f"✅ Converted: {file_path}")
    except Exception as e:
        print(f"❌ Conversion failed for {file_path}: {e}")

# Loop through all .xls files
for file in os.listdir(input_dir):
    if file.endswith(".xls"):
        convert_html_xls_to_xlsx(os.path.join(input_dir, file), converted_dir)
```
